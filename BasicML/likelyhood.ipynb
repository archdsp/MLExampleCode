# 파라미터 공간
## 파라미터
y=ax +b라는 함수가 있다. a와 b가 실수 공간 $\mathbb{R}^2$ 에 위치할 때 실수 공간안에 a와 b가 위치하는 공간을 파라미터 공간이라고 한다. 파라미터는 한국말로 매개변수라고 한다. 변수와 아직 관계가 정해지지 않은 실험을 통해 결정되는 변수로써 변수간의 관계를 표현해주는 변수이다. y = ax + b에서 a와 b값에 따라 x와 y의 관계가 달라진다. a와 b가 x와 y간의 관계를 매개하고 있는 것이다.
a와 b가 각각 1과 2라면 x+2로 2와 3이라면 2x+3으로 x와 y는 x에서 2 더한 관계였다가 두 배하고 3을 더한 관계로 변할 수 있게 된다.

코드에서는 function (a, b) 에서 입력을 받는 변수 a와 b가 된다. 이 또한 입력인 argument를 입력으로 받아 매개하기 때문이다.  
function(1, 2) 에서 1과 2는 argument 인자가 된다. 그리고 function의 a와 b는 a=1 b=2로 인자를 받는 매개변수가 된다. 매개에   의미에 맞춰 인자를 매개했기 때문이다. 원인을 매개했다. 이렇게 생각해본적이 있었나.  

## 매개하다라는 뜻과 parameter, argument
parameter와 매개하다는 둘이 서로 다른 말이지만 정말 잘 표현됐다고 생각한다. parameter는 무언가를 제어하기 위한 제한점이라고   번역되고 우리말 매개하다는 관계를 맺어준다는 뜻이있다. 어느쪽이던 정말 잘 지은 것 같다. argument가 하나가 왜 인자를 말하는 것인지  모르겠다. 영문은 논쟁 논의를 뜻하면서 한국어는 인자 원인을 유발하는 것으로 쓰였다. 논쟁과 논의거리가 어떤 결과에 원인으로 출발선에  있긴 하지만 정말 무슨 뜻인지 궁금하다.

## 공간
만일 a와 b가 아래 그림처럼 1,0 점을 기준점(평균)으로 삼아 분포를 그려보면 초록색 원처럼 인근 분포를 가지게 될 것이다. 이 분포에서는 ax+b로 사상시켰을 때 그려질 수 있는 선분은 대개 y=x 그래프와 유사할 것이다. 만일 a와 b에 확률 분포가 있다면 ax+b가 어떤 일차함수에 근사할지도 알아낼 수 있을 것 같다.

![](https://velog.velcdn.com/images%2Fqsdcfd%2Fpost%2Fd9a116c8-4702-40fd-8e1e-a58b09398f63%2Fimage.png)

이 [글](https://zzaebok.github.io/machine_learning/bayesian-statistics-and-machine-learning/)과 이 [글](https://gentlej90.tistory.com/61)에서는 베이즈 정리와  머신러닝의 관계에 대해서 표현하고 있는데, 나는 기존에 그저 오차를 줄이는 방식으로 최적의 파라미터를 찾아내는 것이 ML이다 생각했었는데  파라미터에 확률분포가 있을 수 있고 우리가 계산하고자 하는 것이 데이터를 통해 현상을 잘 설명해내는 파라미터 분포를 찾는   것이라는 관점은 생각해보지 못했다. 베이즈 정리가 어떻게 ML과 관련이 있는지 아래에서 이어 서술하여 둘간 무슨 관계가 있는지  알아보려고 한다.

### 간단 요약
[글](https://zzaebok.github.io/machine_learning/bayesian-statistics-and-machine-learning/) 에서는 선형회귀에서 찾고자 하는 a와 b가 확률 분포를 갖는다고 했을 때 아래 수식에서 나타내듯 데이터와 모델의 조건부 확률로 표현될 수 있다. 수식에 따르면 사전확률(prior)과 우도(likelyhood, 반대의 조건부확률)확률을 알면 사후확률을 알 수 있게 된다. 모델을 학습시킨다는 것은 관측된 데이터를 이용해 사후확률 얻고 사후확률의 변경에 따라 사전확률(모델의 에측)이 업데이트가 되면서 업데이트 되는 사전확률(모델 예측)를 찾아내게 된다. 그리고 이 모델은 ax+b 선형으로 표현되었을 때 우리가 기대하는 분포를 갖게 만드는 a와 b를 찾아내게 된다.

$$P(Model|Data) = \frac{P(Data|Model)P(Model)}{P(Data)}$$

만약 로지스틱 회귀로 스팸메일 분류를 했다면, 스팸으로 분류한 사건이 있고, 모델 또한 스팸으로 분류한 케이스가 있을 것이다. 데이터를 이용하여 P(Model)값을 업데이트해가며 이 P(Model)이 높아지게 만드는 a와 b의 분포를 찾아내면 된다.

아직까지 정확히 어떻게 a와 b를 찾아낼 것인지는 잘 모르겠다. 무작위로 많이 던져서 점을 찍어서 육안으로 봐야지 하는 상상은 했어도.

# 사전확률(Prior), 우도(Likelyhood), 사후확률(Posterior)
위의 P(Model)은 P(H)의 분포를 의미한다. 머신러닝의 목표는 어떤 데이터집합 $X=\{x_0, x_1, \cdots , x_n\}$의 분포 P(X)를 가장 잘 표현하는 P(W)를 찾는 것이다.
$$
H = \begin{bmatrix} a \\ b \end{bmatrix}
$$

P(X=x|H)는 W가 결정되었을 때 x가 일어날(관측될) 확률로 여기서 x는 고정된 값인 이미 가지고 있는 데이터를 의미한다. 즉 H(a,b)가 정해졌을 때 x를 관측할 수 있느냐? 맟췄느냐를 물어보게 된다. 이 확률을 likelyhood(가능도, 우도)라고 한다. 우도가 높아진다는 얘기는 H가 정해진 조건하에서 데이터 x가 관찰될 확률이 높다는 것을 의미한다. 즉 x의 분포를 잘 표현해낼 수 있다는 것을 의미한다.

이 우도가 높아질 수 있도록 W를 업데이트하여 학습하는 방법을 [최대가능도 추정(Maximum likelyhood estimation)](#최대가능도-추정(Maximum-likelyhood-estimation))이라고 한다.

## 사후확률(Posterior)
우도와 반대의 조건부확률로 P(H|X) 이미 관측된 데이터 X의 확률분포에 따른 가중치 W의 분포라서 사후확률이라 부른다. 이미 사건은 일어났기 때문에. 이 사후확률이 높아질 수 있도록 학습하는 방법을 [최대 사후확률 추정(maximum a posteriori estimation)](#최대-사후확률-추정(maximum-a-posterior-estimation))이라고 한다.

## 우도(Likelihood)
확률(Probability)은 모수안에서 관찰된 사건의 비라면 우도는 반대로 관찰된 사건을 보고 모수가 무엇인지를 찾으려고 할 때 보게되는 확률이다. 그래서 최대우도법 MLE는 우도를 가장 높게 만들어주는 모수를 거꾸로 추정해내는 것이다. 

# 베이즈 정리
위에서 등작한 아래 수식을 베이즈 정리라고 한다.

$$P(Model|Data) = \frac{P(Data|Model)P(Model)}{P(Data)}$$

데이터의 확률 변수 X와 모델 H의 결합 확률(joint probability)는 아래와 같다.  
P(X, H) = P(H|X)P(X) = P(X|H)P(H)  

결합확률은 두 개 이상의 확률 변수를 고려하는(다변수함수, n차원 함수) 확률변수함수를 의미한다. 조건부확률의 분모에 해당하는 확률을 구하니까 결국 $H \cap X$ 확률을 구하는게 된다. 그니까 모델도 맞고 데이터도 맞고. 저 결합확률 식에서 P(H|X)에 관해 정리해주면 아래와 같은 베이즈 정리가 나온다.

$$P(H|X) = \frac{P(X|H)P(H)}{P(X)}$$

## 베이즈 정리와 머신러닝의 관계
정답 데이터를 Y라고 한다면 우리가 원하는 것은 H=(a, b) 가 결정되었을 때 H와 X 결합확률 분포에 따는 Y 즉, 정답이 나타날 확률  
P(Y|W,X) 우도값이다.

아래 그림은 만약 오차(노이즈)가 정규 분포를 따른다면,

![](https://www.sfu.ca/sonic-studio-webdav/handbook/Graphics/Gaussian.gif)

모델의 예측값에다가 노이즈를 더하면 정답 Y값을 구할 수 있으므로 오차가 정규분포를 따른 다는 가정하에 정답이 나타날 확률(우도) 분포를 나타낸 것이다. 이 때 x, y평면에서 구하고자 하는 Wx+b 선분이 지나간다.
가로축은 모델의 예측, 세로축은 P(실제 정답 Y| H,X)의 값이다. 따라서 값이 가장 큰게 좋은데 정규분포를 따른다면 중앙에 가장 높이 봉우리가 있는 지점이므로 해당 지점은 $x_n$에 대응되는 $y_n$으로 가로픅 $H \cdot x_n$ 예측값과 대응되는 출력값이 같을 때 가장 크게 된다.

![](https://velog.velcdn.com/images%2Fqsdcfd%2Fpost%2Fb21ac29d-744f-4de5-9b88-adbe2e22de57%2Fimage.png)

아래 코드는 가우시안 노이즈를 이용해서 편차가 0.1일 때 가우시안 노이즈(정규분포 따르는 오차, 노이즈)를 그린 것이다. 노이즈가 정규분포를 따르면 출력값 또한 정규분포를 따르게 된다. 이때 우리가 다만 우도 확률분포는 평균값이 $H^{T}x$ 일 뿐.

코드는 우도 확률분포 함수를 통해 출력값 별 우도를 계산하여 점 옆에 표시하고 있다. 이 우도값을 보면 임의로 선정한 선형모델 일차함수로부터 멀수록 우도가 낮고 함수 위에 있을 수록 우도가 높아진다.

```python
# model: y = ax + b

import math
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(321)

input_data = np.linspace(-2, 2, 5)
label = input_data + 1 + np.random.normal(0, 1, size=5)

a = 1
b = 2

# 모델 예측값
model_output = a*input_data + b
likelihood = []

# x: 입력 데이터, y: 데이터 라벨
# 예측값과 라벨의 차이를 제곱하여 exp에 사용
for x, y, output in zip(input_data, label, model_output):
    likelihood.append(1/(math.sqrt(2 * math.pi * 0.1 * 0.1))*math.exp(-pow(y - output, 2) / (2 * 0.1 * 0.1)))

model_x = np.linspace(-2, 2, 50)
model_y = a*model_x + b

fig, ax = plt.subplots()
ax.scatter(input_data, label)
ax.plot(model_x, model_y)

for i, text in enumerate(likelihood):
    ax.annotate('%.3e'%text, (input_data[i], label[i]))

plt.show()
```
