```python
import tensorflow as tf
from keras import layers, Input, Model, optimizers
import numpy as np

tf.keras.utils.set_random_seed(0)

# MNIST 데이터 로드
(x_train, y_train), (_, _) = tf.keras.datasets.mnist.load_data()
x_train = (x_train.astype(np.float32) - 127.5) / 127.5  # Normalize [-1, 1]
x_train = np.expand_dims(x_train, axis=-1)
y_train = y_train.reshape(-1, 1)

batch_size = 512

# 데이터셋 준비
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(buffer_size=128).batch(batch_size)

latent_dim = 100
num_classes = 10

def build_generator():
    z = Input(shape=(latent_dim, 1))
    labels = Input(shape=(1,))

    # Embedding layer to transform labels
    embedding = layers.Embedding(input_dim=num_classes, output_dim=num_classes)(labels)
    embedding = layers.Reshape((num_classes, 1))(embedding)

    x = layers.Concatenate(axis=1)([z, embedding])
    x = layers.Flatten()(x)
    x = layers.Dense(7 * 7 * 128, activation='relu')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.Reshape((7, 7, 128))(x)

    x = layers.Conv2DTranspose(128, kernel_size=3, strides=2, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2DTranspose(64, kernel_size=3, strides=2, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2DTranspose(32, kernel_size=3, strides=1, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2DTranspose(1, kernel_size=3, strides=1, padding='same', activation='tanh')(x)

    return Model([z, labels], x)


generator = build_generator()


def build_discriminator():
    image = Input(shape=(28, 28, 1))
    labels = Input(shape=(1,))

    embedding = layers.Embedding(input_dim=num_classes, output_dim=(28 * 28))(labels)
    embedding = layers.Reshape((28, 28, 1))(embedding)

    x = layers.Concatenate(axis=-1)([image, embedding])

    x = layers.Conv2D(32, kernel_size=3, strides=2, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2D(64, kernel_size=3, strides=2, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Conv2D(128, kernel_size=3, strides=2, padding='same')(x)
    x = layers.LeakyReLU(alpha=0.2)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(1, activation="sigmoid")(x)

    return Model([image, labels], x)

discriminator = build_discriminator()

class CGAN(Model):
    def __init__(self, generator, discriminator):
        super(CGAN, self).__init__()
        self.generator = generator
        self.discriminator = discriminator

    def compile(self, d_optimizer, g_optimizer, loss_fn):
        super(CGAN, self).compile(run_eagerly=True)
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn
    
    def train_step(self, data):
        real_images, real_labels = data
        batch_size = tf.shape(real_images)[0]
        random_latent_vectors = tf.random.normal(shape=(batch_size, latent_dim))
        random_labels = tf.random.uniform(shape=(batch_size, 1), minval=0, maxval=num_classes, dtype=tf.int32)

        # 판별자 학습
        with tf.GradientTape() as tape:
            generated_images = self.generator([random_latent_vectors, random_labels], training=True)
            fake_output = self.discriminator([generated_images, random_labels], training=True)
            real_output = self.discriminator([real_images, real_labels], training=True)

            # 진짜 이미지에 대한 손실 (레이블 = 1)
            d_loss_real = self.loss_fn(tf.ones_like(real_output), real_output)
            # 가짜 이미지에 대한 손실 (레이블 = 0)
            d_loss_fake = self.loss_fn(tf.zeros_like(fake_output), fake_output)
            d_loss = (d_loss_real + d_loss_fake) / 2

        d_gradients = tape.gradient(d_loss, self.discriminator.trainable_variables)
        self.d_optimizer.apply_gradients(zip(d_gradients, self.discriminator.trainable_variables))

        # 생성자 학습
        with tf.GradientTape() as tape:
            generated_images = self.generator([random_latent_vectors, random_labels], training=True)
            fake_predictions = self.discriminator([generated_images, random_labels], training=True)
            # 생성자는 가짜 이미지를 진짜로 판별하도록 학습 (레이블 = 1)
            g_loss = self.loss_fn(tf.ones_like(fake_predictions), fake_predictions)

        g_gradients = tape.gradient(g_loss, self.generator.trainable_variables)
        self.g_optimizer.apply_gradients(zip(g_gradients, self.generator.trainable_variables))

        return {"d_loss": d_loss, "g_loss": g_loss, "d_real_loss": d_loss_real, "d_fake_loss": d_loss_fake}
    

import matplotlib.pyplot as plt

class ImageGeneratorCallback(tf.keras.callbacks.Callback):
    def __init__(self, generator, latent_dim, num_classes, num_samples=5):
        self.generator = generator
        self.latent_dim = latent_dim
        self.num_classes = num_classes
        self.num_samples = num_samples

    def on_epoch_end(self, epoch, logs=None):
        # Generate random latent vectors and integer labels
        random_latent_vectors = tf.random.normal(shape=(self.num_samples, self.latent_dim, 1))
        random_labels = np.random.randint(0, self.num_classes, self.num_samples).reshape(-1, 1)

        # Convert random_labels to a TensorFlow tensor
        random_labels = tf.convert_to_tensor(random_labels, dtype=tf.int32)

        # Generate images using the generator
        generated_images = self.generator([random_latent_vectors, random_labels], training=False)

        # Plot the images
        _, axes = plt.subplots(1, self.num_samples, figsize=(15, 15))
        for i in range(self.num_samples):
            ax = axes[i]
            img = generated_images[i, :, :, 0] * 127.5 + 127.5
            ax.imshow(img, cmap='gray')  # Assumes grayscale images
            ax.axis('off')
            ax.set_title(f"Label: {random_labels[i, 0].numpy()}")
        plt.show()

d_optimizer = optimizers.Adam(learning_rate=0.0003, beta_1=0.5)
g_optimizer = optimizers.Adam(learning_rate=0.0001, beta_1=0.5)

loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=True)
model = CGAN(generator, discriminator)
model.compile(d_optimizer, g_optimizer, loss_fn)

callback = ImageGeneratorCallback(generator, latent_dim, num_classes)
model.fit(train_dataset, epochs=100, callbacks=[callback])
```
